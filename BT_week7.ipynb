```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn import linear_model

#Lấy dữ liệu của file csv
url='1.01. Simple linear regression.csv'
dataframe = pd.read_csv(url)
x = dataframe.set_index(dataframe['SAT'])
y = dataframe.set_index(dataframe['GPA'])

#Vẽ
plt.scatter(x,y)
plt.xlabel('SAT',fontsize = 60)
plt.xlabel('GPA',fontsize = 20)
dataframe.hist()
plt.show()

#sSự tương quan (dương là tương quan dương)
conrrelusion = dataframe.corr()
print(conrrelusion)

#Hồi quy tuyến tính
regression_model = linear_model.LinearRegression()
regression_model.fit(x,y)

print(regression_model.intercept_)
print(regression_model.coef_)
```

```python
#Lấy dữ liệu của file csv
url='mtcars.csv'
data = pd.read_csv(url)

#Sự tương quan (dương là tỷ lệ thuận) từ đó xóa đi cột không cần thiết
conrrelusion = data.corr()
print(conrrelusion)

#Xóa cột 0 tới 2
x = data.drop(data.columns[0:2],axis = 1)
data.plot(kind = "scatter",x = "wt", y = "mpg", figsize = (9,9),color = "black")


```

```python
# Tạo biến dựa trên những atribute của data
X = pd.DataFrame(data['wt'])
y = data['mpg']

#Tạo một model theo dạng hàm bậc 1
model = linear_model.LinearRegression()
#Cho phép giá trị cảu data được đưa về dạng bậc 1 tuyến tính
model.fit(X,y)
print(model.intercept_)
print(model.coef_)
# độ chính xác
model.score(X,y)

y_pred = model.predict(X)
residual = data['mpg'] - y_pred
residual.describe() 

data.plot(kind = "scatter",x = "wt", y = "mpg", figsize = (9,9),color = "black")
plt.plot(data['wt'],y_pred,color = 'red')
```

```python
model = linear_model.LinearRegression()

X = pd.DataFrame([data['wt'],data['wt']**2]).T  #Chuyển vị

model.fit(X,y = data['mpg'])

print('Intercept của hình bậc 2')
print(model.intercept_)
print('Coefiction của hình bậc 2')
print(model.coef_)
print('Độ chính xác của các hình')
print(model.score(X,y))
```

```python
range = np.arange(1.5, 5.5, 0.1)
predictor = pd.DataFrame([range,range**2]).T

y_pred = model.predict(X = predictor)
data.plot(kind = 'scatter',x = 'wt', y = 'mpg', figsize = (9,9), color = 'blue', xlim = (0,7) )
plt.plot(range, y_pred, color = 'red')
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sas

data = pd.read_csv('50_Startups.csv')
data.head()
```

```python
sas.pairplot(data)
```

```python
corr = data.corr()
sas.heatmap(corr,annot = True)
```

```python
X=data.drop(data.columns[[3,4]],axis=True)
y = data['Profit']

X.head()
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X =sc.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2, random_state  = 0)
X_train.shape
X_test.shape
y_train.shape
y_test.shape
```

```python

from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train,y_train)
y_pred = regressor.predict(X_test).round(1)
calculation = pd.DataFrame(np.c_[y_test,y_pred],columns=['Original Profit','Predicted Profit'])

print('Độ chính xác của mô hình')
print(regressor.score(X_train,y_train))
print(regressor.score(X_test,y_test))
```
